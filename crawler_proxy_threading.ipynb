```python
import threading, time, requests, os, queue
import pandas as pd
from bs4 import BeautifulSoup

headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/81.0.4044.138 Safari/537.36'}

def crawler(headers):
    resp = requests.get('https://free-proxy-list.net', headers=headers, timeout=5)
    soup = BeautifulSoup(resp.text, 'lxml')
    trs = soup.find('table').select('tr')
    ip_port_list=[]
    for i in range(1,len(trs)-1):
        tds=trs[i].find_all('td')
        if tds[6].text=='yes':
            ip_port=tds[0].text+':'+tds[1].text
            ip_port_list.append(ip_port)
    return ip_port_list
ip_port_list=crawler(headers)
print(ip_port_list)
print(len(ip_port_list))
```
